## Part 1: Michaelis-Menten equation

```python
# Plot the Michaelis-Menten Equation for Km = 1 and Vmax = 25

Ss = np.linspace(0, 50, 100)

# Parameters to test. First index is Vmax, second index Km
parameters = (
    (25, 1), 
    (25, 0.5),
    (15, 1),
)

for params in parameters:
    Vmax = params[0]
    Km = params[1]

    rates = michaelis_menten_equation(Ss, Vmax, Km)

    plt.plot(Ss, rates, label=f"Vmax={Vmax}, Km={Km}")

plt.xlabel("[S]")
plt.ylabel("v")
plt.legend()
plt.show()
```
